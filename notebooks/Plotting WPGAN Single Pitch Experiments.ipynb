```python
import json
import matplotlib.pyplot as plt
```

```python
with open('../training_history/wpgan_flute_hist_linear_pitch_0729.json', 'r') as fp:
    dcgan_history = json.load(fp)

print([key for key in dcgan_history])
```

```python
plt.plot(dcgan_history['disc_loss'], label="Discriminator")
plt.plot(dcgan_history['gen_loss'], label="Generator")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.savefig('./figures/wpgan_flute_linear_pitch_loss_0729.png', dpi=300)
```

```python
plt.plot(dcgan_history['disc_real_score'], label="Real")
plt.plot(dcgan_history['disc_fake_score'], label="Generated")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.savefig('./figures/0725_hybrid_wpgan_accuracy.png', dpi=300)
```
