```python
import json
import matplotlib.pyplot as plt
```

```python
with open('../training_history/wpgan_flute_hist_upsample_0726.json', 'r') as fp:
    history_1 = json.load(fp)

print([key for key in history_1])
```

```python
plt.plot(history_1['disc_loss'], label="Discriminator")
plt.plot(history_1['gen_loss'], label="Generator")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Loss")
#plt.savefig('./figures/0725_flute_dcgan_loss.png', dpi=300)
```

```python
with open('../training_history/wpgan_flute_hist_upsample_0727.json', 'r') as fp:
    history_2 = json.load(fp)
```

```python
plt.plot(history_2['disc_loss'], label="Discriminator")
plt.plot(history_2['gen_loss'], label="Generator")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Loss")
#plt.savefig('./figures/0725_flute_dcgan_loss.png', dpi=300)
```

```python
full_loss_gen = history_1['gen_loss'] + history_2['gen_loss']
full_loss_disc = history_1['disc_loss'] + history_2['disc_loss']
```

```python
plt.plot(full_loss_disc, label="Discriminator")
plt.plot(full_loss_gen, label="Generator")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.savefig('./figures/0728_flute_upsample_loss.png', dpi=300)
```
