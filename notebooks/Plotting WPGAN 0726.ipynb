```python
import json
import matplotlib.pyplot as plt
```

```python
with open('../training_history/dcgan_flute_hist_0725.json', 'r') as fp:
    dcgan_history = json.load(fp)

print([key for key in dcgan_history])
```

```python
plt.plot(dcgan_history['disc_loss'], label="Discriminator")
plt.plot(dcgan_history['gen_loss'], label="Generator")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.savefig('./figures/0725_flute_dcgan_loss.png', dpi=300)
```

```python
plt.plot(dcgan_history['disc_real_accuracy'], label="Real")
plt.plot(dcgan_history['disc_fake_accuracy'], label="Generated")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.savefig('./figures/0725_flute_dcgan_accuracy.png', dpi=300)
```
